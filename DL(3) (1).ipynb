```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import BatchNormalization, LayerNormalization

# Generate random dataset
x = np.random.randint(1, 100, size=(5, 4)).astype(np.float32)

print("Original Input:")
print(x)

# Batch Normalization
batch_norm = BatchNormalization()
batch_output = batch_norm(x, training=True)

print("\nBatch Normalization Output:")
print(batch_output.numpy())

# Layer Normalization
layer_norm = LayerNormalization()
layer_output = layer_norm(x)

print("\nLayer Normalization Output:")
print(layer_output.numpy())

print("\nComparison:")
print("1. Batch Normalization normalizes data using batch statistics.")
print("2. Layer Normalization normalizes each sample independently.")
print("3. Batch Normalization depends on batch size, while Layer Normalization does not.")
```
